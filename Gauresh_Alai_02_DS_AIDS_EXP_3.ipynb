```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
```

```python
df = pd.read_csv('Car data.csv')
df
```

Mean of Data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df['Insurance'] = imputer.fit_transform(df[['Insurance']])
print(df['Insurance'])
```

Median of data

```python
imputer = SimpleImputer(strategy='median')
df['Insurance']=imputer.fit_transform(df[['Insurance']])
print(df['Insurance'])
```

Label Encoding: Convert categorical labels into numerical integers

```python
label_encoder = LabelEncoder()
df['Insurance'] = label_encoder.fit_transform(df['Insurance'])
print(df['Insurance'])
```

One-Hot Encoding: Create binary columns for each category

```python
ohe = OneHotEncoder()
encoded = ohe.fit_transform(df[['Insurance']])
print(df['Insurance'])
```

```python
features = df.columns.tolist()
# Plot box plots for each feature
plt.figure(figsize=(20, 15))
for i, feature in enumerate(features):
    plt.subplot(6, 4, i + 1)
    sns.boxplot(data=df[feature])
    plt.title(f'Box plot of {feature}')
plt.tight_layout()
plt.show()
```

removing outliers

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
df
```

checking if outliers are removed

```python
features = df.columns.tolist()
# Plot box plots for each feature
plt.figure(figsize=(20, 15))
for i, feature in enumerate(features):
    plt.subplot(6, 4, i + 1)
    sns.boxplot(data=df[feature])
    plt.title(f'Box plot of {feature}')
plt.tight_layout()
plt.show()
```
